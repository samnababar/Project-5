```python
import random

def get_valid_word(words):
    """Select a random word without hyphens or spaces."""
    word = random.choice(words)
    while '-' in word or ' ' in word:
        word = random.choice(words)
    return word.upper()

def hangman():
    word_list = [
        'PYTHON', 'JAVA', 'JAVASCRIPT', 'RUBY', 'SWIFT',
        'PROGRAMMING', 'DEVELOPER', 'ALGORITHM', 'FUNCTION',
        'VARIABLE', 'STRING', 'INTEGER', 'BOOLEAN', 'HANGMAN'
    ]
    word = get_valid_word(word_list)
    word_letters = set(word)
    alphabet = set('ABCDEFGHIJKLMNOPQRSTUVWXYZ')
    used_letters = set()
    lives = 6

    print("ğŸ® **Welcome to Hangman!**")
    print(f"Guess the word (Hint: It has {len(word)} letters)")

    while len(word_letters) > 0 and lives > 0:
        print("\nğŸ’€ Lives left:", lives)
        print("ğŸ”  Used letters:", " ".join(sorted(used_letters)))

        # Display word with blanks (e.g., _ _ P _ _ N)
        display_word = [letter if letter in used_letters else '_' for letter in word]
        print("ğŸ“œ Word:", " ".join(display_word))

        # Get user input
        user_letter = input("ğŸ” Guess a letter: ").upper()

        if user_letter in alphabet - used_letters:
            used_letters.add(user_letter)
            if user_letter in word_letters:
                word_letters.remove(user_letter)
                print("âœ… Correct!")
            else:
                lives -= 1
                print("âŒ Wrong guess!")
        elif user_letter in used_letters:
            print("âš ï¸ You already guessed that letter!")
        else:
            print("ğŸš« Invalid input! Use A-Z only.")

    # Game over message
    if lives == 0:
        print(f"\nâ˜ ï¸ **GAME OVER!** The word was: {word}")
    else:
        print(f"\nğŸ‰ **YOU WIN!** The word was: {word}")

# Start the game
hangman()
```
